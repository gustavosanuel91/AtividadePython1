```python
usuario = "Gustavo"
senha = 1234

nome_usuario = str(input("Digite seu nome: "))
senha_login = int(input("Digite sua senha: "))

if nome_usuario == usuario and senha_login == senha:
    print("\n=====| Login realizado com sucesso! |=====")
else:
    print("\n=====| Usuário ou senha incorretos |=====")
```
